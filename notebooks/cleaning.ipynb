```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw.csv")
df.head(5)
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
from sklearn.preprocessing import OneHotEncoder

cols = ['Airline','Source','Destination']
encoder = OneHotEncoder()

encoded_data = encoder.fit_transform(df[cols]).toarray()
encoded_df = pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out())
```

```python
df = pd.concat([df,encoded_df],axis=1)
```

```python
df.drop(cols,axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df['date'] = df['Date_of_Journey'].str.split("/").str[0]
df['month'] = df['Date_of_Journey'].str.split("/").str[1]
df['year'] = df['Date_of_Journey'].str.split("/").str[2]
```

```python
df.drop(['Date_of_Journey'],axis = 1,inplace=True)
```

```python
df.head(5)
```

```python
df.drop(['Route'],axis = 1,inplace=True)
```

```python
df['dep_hour'] = df['Dep_Time'].str.split(":").str[0]
df['dep_min'] = df['Dep_Time'].str.split(":").str[1]
```

```python
df.drop(['Dep_Time'],axis = 1,inplace=True)
```

```python
df.head(5)
```

```python
df['Arrival_hr'] = df['Arrival_Time'].str.split(":").str[0]
df["Arrival_min"] = df['Arrival_Time'].str.split(":").str[1].str.split(" ").str[0]
```

```python
df.drop(['Arrival_Time'],axis = 1,inplace=True)
```

```python
df.head(1)
```

```python
df['Duration_hr'] = df['Duration'].str.split(" ").str[0].str.split("h").str[0]
df['Duration_min'] = df['Duration'].str.split(" ").str[1].str.split("m").str[0]
```

```python
df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
df.drop(['Duration'],axis = 1,inplace=True)
```

```python
df.head(5)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].replace({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,np.nan:0})
```

```python
df.head(5)
```

```python
df.drop(['Additional_Info'],axis = 1,inplace=True)
```

```python
df.to_csv('cleaned.csv',index=False)
```
